```python
import pygame, time, random
GRAY=(119,118,110)
BLACK=(0,0,0)
RED=(255,0,0)
GREEN=(0,200,0)
BLUE=(0,0,200)
BRIGHT_RED=(255,0,0)
BRIGHT_GREEN=(0,255,0)
BRIGHT_BLUE=(0,0,255)
SCREEN_WIDTH=800
SCREEN_HEIGHT=600

class UI():
    def __init__(self):
        font = 'freesansbold.ttf'
        self.xltext=pygame.font.Font(font,115)
        self.lgtext=pygame.font.Font(font,80)
        self.smtext=pygame.font.Font(font,20)
        self.mdtext=pygame.font.Font(font,40)

        self.backgroundpic=pygame.image.load("download12.jpg")
        self.yellow_strip=pygame.image.load("yellow strip.jpg")
        self.strip=pygame.image.load("strip.jpg")

    def message(self, text, size = 'lg', color = BLACK, center = (SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2)):
        textsurface=getattr(self, size + "text" ).render(text,True,color)
        r = textsurface.get_rect()
        r.center = center
        gamedisplays.blit(textsurface,r)

    def button(self,msg,x,y,w,h,ic,ac,action):
        mouse=pygame.mouse.get_pos()
        click=pygame.mouse.get_pressed()
        if x+w>mouse[0]>x and y+h>mouse[1]>y:
            pygame.draw.rect(gamedisplays,ac,(x,y,w,h))
            if click[0]==1:
                action()
        else:
            pygame.draw.rect(gamedisplays,ic,(x,y,w,h))

        self.message(msg, 'sm', BLACK, ((x+(w/2)),(y+(h/2))))

    def background(self, rel_y):
        gamedisplays.fill(GRAY)
        gamedisplays.blit(self.backgroundpic,(0,rel_y-self.backgroundpic.get_rect().width))
        gamedisplays.blit(self.backgroundpic,(700,rel_y-self.backgroundpic.get_rect().width))
        if rel_y<800:
            gamedisplays.blit(self.backgroundpic,(0,rel_y))
            gamedisplays.blit(self.backgroundpic,(700,rel_y))
            for _y in [0, 100, 200, 300, 400, 500, -100]:
                gamedisplays.blit(self.yellow_strip,(400,rel_y + _y))
            gamedisplays.blit(self.strip,(120,rel_y-200))
            gamedisplays.blit(self.strip,(120,rel_y+20))
            gamedisplays.blit(self.strip,(120,rel_y+30))
            gamedisplays.blit(self.strip,(680,rel_y-100))
            gamedisplays.blit(self.strip,(680,rel_y+20))
            gamedisplays.blit(self.strip,(680,rel_y+30))




class Car():
    def __init__(self):
        self.image = pygame.image.load('car1.jpg')
        self.width = 56
        self.x = 0
        self.y = 0
        self.xdelta = 0

    def draw(self):
        gamedisplays.blit(self.image,(self.x,self.y))

    def check_outbound(self):
        if self.x>690-self.width or self.x<110:
            return True
        if self.x>SCREEN_WIDTH-(self.width+110) or self.x<110:
            return True
        return False

class Obstacle():
    def __init__(self):
        self.images =  ["car.jpg", "car1.jpg", "car2.jpg", "car4.jpg", "car5.jpg", "car6.jpg", "car7.jpg"]
        self.images = [pygame.image.load(x) for x in self.images]
        self.mode = 0
        self.x = 0
        self.y = 0
        self.width=56
        self.height=125
        self.speed=0

    def draw(self):
        gamedisplays.blit(self.images[self.mode],(self.x,self.y))

    def check_crash(self, car):
        if car.y<self.y+self.height:
            if car.x > self.x and car.x < self.x + self.width or car.x+car.width > self.x and car.x+car.width < self.x+self.width:
                return True
        return False

pygame.init()
gamedisplays=pygame.display.set_mode((SCREEN_WIDTH,SCREEN_HEIGHT))
pygame.display.set_caption("Car Game")
clock=pygame.time.Clock()
ui = UI()
car = Car()
obs = Obstacle()
pause=False

def intro_loop():
    intro_background=pygame.image.load("background.jpg")
    intro=True
    while intro:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                shutdown()
        gamedisplays.blit(intro_background,(0,0))
        ui.message('CAR GAME', 'xl', center=(400,100))
        ui.button("MULAI",150,520,100,50,GREEN,BRIGHT_GREEN,countdown)
        ui.button("KELUAR",550,520,100,50,RED,BRIGHT_RED,shutdown)
        ui.button("CARA BERMAIN",300,520,200,50,BLUE,BRIGHT_BLUE,introduction)
        pygame.display.update()
        clock.tick(50)

def shutdown():
    pygame.quit()
    quit()
    sys.exit()

def introduction():
    instruction_background=pygame.image.load("background2.jpg")
    introduction=True
    while introduction:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                shutdown()
        gamedisplays.blit(instruction_background,(0,0))

        ui.message("Permainan mobil di mana Anda perlu menghindari mobil yang datang", 'sm', center = (400, 200))
        ui.message("CARA BERMAIN", 'lg', center = (400, 100))

        ui.message(" KIRI : BELOK KIRI", 'sm', center = (150, 400))
        ui.message(" KANAN : BELOK KANAN", 'sm', center = (150, 450))
        ui.message("A : GAS ", 'sm', center = (150, 500))
        ui.message("B : REM ", 'sm', center = (150, 550))
        ui.message("P : PAUSE  ", 'sm', center = (150, 350))
        ui.message("CONTROLS", 'md', center = (350, 300))

        ui.button("KEMBALI",600,450,100,50,GREEN,BRIGHT_GREEN,intro_loop)
        pygame.display.update()
        clock.tick(30)

def paused():
    global pause
    instruction_background=pygame.image.load("background2.jpg")
    while pause:
            for event in pygame.event.get():
                if event.type==pygame.QUIT:
                    pygame.quit()
                    quit()
                    sys.exit()
            gamedisplays.blit(instruction_background,(0,0))
            ui.message('PAUSED', 'xl')
            ui.button("LANJUT",150,450,150,50,GREEN,BRIGHT_GREEN,unpaused)
            ui.button("ULANG",350,450,150,50,BLUE,BRIGHT_BLUE,countdown)
            ui.button("MENU",550,450,200,50,RED,BRIGHT_RED,intro_loop)
            pygame.display.update()
            clock.tick(30)

def unpaused():
    global pause
    pause=False

def countdown():
    countdown=True

    while countdown:
            for event in pygame.event.get():
                if event.type==pygame.QUIT:
                    pygame.quit()
                    quit()
                    sys.exit()

            for c in ["3", "2", "1", "AYO!!!"]:
                gamedisplays.fill(GRAY)
                ui.background(0)
                car.x=(SCREEN_WIDTH*0.45)
                car.y=(SCREEN_HEIGHT*0.8)
                car.draw()
                ui.message(c, 'xl')
                pygame.display.update()
                clock.tick(1)

            game_loop()

def score_system(passed,score):
    font=pygame.font.SysFont(None,25)
    text=font.render("MENYALIP %d" % (passed),True,BLACK)
    score=font.render("SCORE %d" % (score),True,RED)
    gamedisplays.blit(text,(0,50))
    gamedisplays.blit(score,(0,30))

def crash():
    ui.message("WOW NABRAK!")
    pygame.display.update()
    time.sleep(3)
    game_loop()

def game_loop():
    global pause
    car.x=(SCREEN_WIDTH*0.45)
    car.y=(SCREEN_HEIGHT*0.8)
    car.xdelta=0
    obs.speed=9
    obs.x = random.randrange(200,(SCREEN_WIDTH-200))
    obs.y = -750
    passed=0
    level=0
    score=0
    y2=7

    bumped=False
    while not bumped:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                shutdown()

            if event.type==pygame.KEYUP:
                if event.key==pygame.K_LEFT or event.key==pygame.K_RIGHT:
                    car.xdelta=0

            if event.type==pygame.KEYDOWN:
                if event.key==pygame.K_LEFT:
                    car.xdelta=-5
                if event.key==pygame.K_RIGHT:
                    car.xdelta=5
                if event.key==pygame.K_a:
                    obs.speed+=2
                if event.key==pygame.K_b:
                    obs.speed-=2

        car.x+=car.xdelta
        pause=True

        rel_y=y2%ui.backgroundpic.get_rect().width
        ui.background(rel_y)

        y2+=obs.speed

        obs.y-=(obs.speed/4)
        obs.draw()
        obs.y+=obs.speed
        car.draw()
        score_system(passed,score)

        if car.check_outbound():
            crash()
        if obs.check_crash(car):
            crash()

        if obs.y>SCREEN_HEIGHT:
            obs.y=0-obs.height
            obs.x=random.randrange(170,(SCREEN_WIDTH-170))
            obs.mode=random.randrange(0,7)
            passed=passed+1
            score=passed*10
            if int(passed)%10==0:
                level=level+1
                obs.speed+2
                ui.message("LEVEL %d" % level, 'lg')
                pygame.display.update()
                time.sleep(3)

        ui.button("Pause",650,0,150,50,BLUE,BRIGHT_BLUE,paused)
        pygame.display.update()
        clock.tick(60)

intro_loop()
game_loop()
pygame.quit()
quit()
```
